```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Explanation of Columns:
1. Country
2. Year - (year data was collected)
3. Status - (Developing or Developed)
4. Life Expectancy - (in age)
5. Adult Mortality - (rates of both sexes probability of dying between 15 and 60 years per 1000 population)
6. Infant deaths - (number of infant deaths per 1000 population)
7. Alcohol - (alchohol consumption recorded per capita in liters of pure alcohol)
8. Percentage expenditure -(expenditure on health as a percentage of domestic product per capita)
9. Hepatitis B - (HepB) immunization coverage among 1 year olds (%)
10. Measles - Number of reported cases per 1000 population
11. BMI - Average body mass index of entire population
12. Under-five deaths - Number of under-five deaths per 1000 population
13. Polio - Polio (Pol3) immunization coverage among 1 year olds
14. Total expenditure - General government expenditure on health as a percentage of total government expenditure (%)
15. Diphteria - Diptheria tetanus toxoid and pertussis (DTP3) immunization covered among 1 year olds (%)
16. HIV/AIDS - Deaths per 1000 live births HIV/AIDS (0-4 years)
17. GDP - Gross Domestic Product per capita (in USD)
18. Population - Population of the country
19. Thinness 1-19 years - Prevalence of thinness among children and adolescents for Ages 10-19 (%)
20. Thinness 5-9 years - Prevalence of thinness among children for Ages 5-9 (%)
21. Income composition - Human Development Index in terms of income composition of resources (index ranging from 0 to 1)
22. Schooling - Number of years of schooling (years)

Dataset: https://www.kaggle.com/kumarajarshi/life-expectancy-who

```python
#LINEAR REGRESSION MODEL
from sklearn.linear_model import LinearRegression
dfDevelop = pd.get_dummies(frame['Status'])
frame2 = pd.concat([frame, dfDevelop], axis=1)
print(frame2)
y = frame2['Life expectancy']
X = frame2.iloc[:, 4:]
print(X.shape)
print(y.shape)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print(y_train.shape)
print(X_train.shape)
print(y_test.shape)
print(X_test.shape)
model = LinearRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
from sklearn.metrics import r2_score
print("R^2 value:",r2_score(y_test, y_pred))
#RESIDUALS PLOT
from yellowbrick.regressor import ResidualsPlot
visualizer = ResidualsPlot(model)
visualizer.fit(X, y)
visualizer.poof()
plt.show()
```

```python
#RANDOM FOREST REGRESSION MODEL
from sklearn.ensemble import RandomForestRegressor
y = frame2['Life expectancy']
X = frame2.iloc[:, 4:]
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
rf = RandomForestRegressor(random_state = 42, n_estimators=100)
rf.fit(X_train, y_train)
predict_rf = rf.predict(X_test)
#ACCURACY SCORE
print("Training data score:", rf.score(X_train, y_train))
print("Testing data score:", rf.score(X_test, y_test))
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
print("R2 Score", r2_score(y_test, predict_rf))
print("Mean Squared Error", mean_squared_error(y_test,predict_rf)**(0.5))
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(rf,X_train,y_train,cv=10)
print(accuracies.mean())
from sklearn.model_selection import RandomizedSearchCV
from pprint import pprint
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 2000, num = 10)]
max_features = ['auto', 'sqrt']
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
min_samples_split = [2, 5, 10]
min_samples_leaf = [1, 2, 4]
bootstrap = [True, False]
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
pprint(random_grid)
rf = RandomForestRegressor()
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
rf_random.fit(X_train, y_train)
```

```python
print(rf_random.best_params_)
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    return accuracy
base_model = RandomForestRegressor(n_estimators = 10, random_state = 42)
base_model.fit(X_train, y_train)
accu_score = []
base_accuracy = evaluate(base_model, X_test, y_test)
accu_score.append(base_accuracy)
best_random = rf_random.best_estimator_
random_accuracy = evaluate(best_random, X_test, y_test)
accu_score.append(random_accuracy)
print('Improvement of {:0.2f}%.'.format( 100 * (random_accuracy - base_accuracy)/ base_accuracy))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'bootstrap': [True, False],
    'max_depth': [10, 20, 50, None],
    'max_features': [2, 3, 4, 'auto'],
    'min_samples_leaf': [1, 3, 5],
    'min_samples_split': [1, 2, 4, 8],
    'n_estimators': [10, 30, 100, 120, 150]
}
rf = RandomForestRegressor()
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 3, n_jobs = -1, verbose = 2)
grid_search.fit(X_train, y_train)
print(grid_search.best_params_)
best_grid = grid_search.best_estimator_
grid_accuracy = evaluate(best_grid, X_test, predict_rf)
print('Improvement of {:0.2f}%.'.format( 100 * (grid_accuracy - base_accuracy) / base_accuracy))
model_predictions = best_grid.predict(X_test)
residuals = model_predictions - predict_rf
from sklearn.metrics import r2_score
print("R-squared Score Obtained for predicting Life Expectancy is {0}".format(r2_score(predict_rf, model_predictions)))
```

```python
grid_accuracy = evaluate(best_grid, X_test, predict_rf)
accu_score.append(grid_accuracy)
```

```python
#ACCURACY BAR CHART
xvalues = ["Base Model", "Random Search", "Grid Search"]
print(accu_score)
plt.bar(xvalues, accu_score, color='r', linewidth=1.5, edgecolor='k')
plt.ylim(ymin=98, ymax=100)
plt.xticks(xvalues, fontsize=14)
plt.xlabel("Model")
plt.ylabel("Accuracy(%)")
plt.title("Accuracy Comparison")
plt.show()
```
